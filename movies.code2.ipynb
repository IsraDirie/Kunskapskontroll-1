```python
import pandas as pd

df = pd.read_json(r"C:\Users\israd\OneDrive\Dokument\TUC\Databastyper\movies (1).json")

movie_id_genres = df[["genres", "movie_id"]].explode("genres").dropna(axis=0)

genres_df = pd.DataFrame(movie_id_genres["genres"].unique(), columns=["genres"])
genres_df["genre_id"] = genres_df.index + 1

merged_df = pd.merge(movie_id_genres, genres_df, on="genres", how="left")

merged_df = merged_df.drop("genres", axis=1)
print(merged_df.head())


merged_df.to_csv("genres_movie_2.csv", index=False)
```
